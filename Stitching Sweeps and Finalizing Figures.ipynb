```python
def init_plot(i):
    i.plot()
    plt.xlim([min(i.x),max(i.x)])
    plt.ylim([min(i.y)-0.05,max(i.y)+0.05])
    
def stitch(i,x):
    i.fits[1].plot_select('b',['k','r'],[2,2],i.fits[1].pos_left,x)
    i.fits[0].plot_select('b',['r','#00ff00'],[2,2],x,i.fits[0].pos_right)
```
