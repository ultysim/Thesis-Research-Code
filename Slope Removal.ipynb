```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import Analysis
import pylab
import regression_cv as rcv
import ipywidgets
%matplotlib 
```

```python
data2011 = np.load('2011_Prune_strong.npy')
```

```python
data_hold = np.array(data2011)
```

```python
filenum = 7
x,y = data2011[filenum][0],data2011[filenum][1]
%matplotlib inline

def func(slope,bias):
    fig = plt.figure(figsize=(13,4))
    slope_subtraction = y - slope*x + bias
    plt.plot(x,slope_subtraction)
    plt.title('Slope subtraction with {} slope and {} bias'.format(slope,bias))
    plt.ylim(min(y)-0.05, max(y)+0.05)
    plt.xlim(min(x),max(x))
    plt.grid(True)
    plt.show()   


interactive_plot = ipywidgets.interactive(func, slope=(-0.005, 0.005, 0.0001), bias=(-0.5, 0.5, 0.01))
output = interactive_plot.children[-1]
output.layout.height = '350px'
output.layout.width = '700px'
interactive_plot
```
